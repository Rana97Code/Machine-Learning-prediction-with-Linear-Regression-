```python
import pandas as pd

# Create a simple DataFrame
data = {
    "StudyHours": [1, 2, 3, 4, 5],
    "Scores": [40, 50, 55, 65, 80]
}

df = pd.DataFrame(data)
df
```

```python
import matplotlib.pyplot as plt

plt.scatter(df["StudyHours"], df["Scores"])
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Study Hours vs Scores")
plt.show()
```

```python
from sklearn.linear_model import LinearRegression

# Reshape the X values
X = df[["StudyHours"]]  # features
y = df["Scores"]        # target

model = LinearRegression()
model.fit(X, y)

model
```

```python
predict_value = model.predict([[6]])
predict_value
```

```python
predict_value = model.predict([[8]])
predict_value
```

```python
plt.scatter(X, y)
plt.plot(X, model.predict(X), linewidth=2)
plt.xlabel("Study Hours")
plt.ylabel("Scores")
plt.title("Linear Regression Prediction")
plt.show()
```
